```python
from BaseImage import BaseImage
import matplotlib.pyplot as plt

img_arr = BaseImage('D:\STUDIES\WDGM\WDGM\lab2\lena.jpg')
base_data = img_arr.data
img_arr.to_hsv()
hsv_data = img_arr.data
img_arr.to_rgb()
rgb_from_hsv = img_arr.data

f, ax_arr = plt.subplots(1, 2)
ax_arr[0].imshow(base_data)
ax_arr[0].set_title("RGB")
ax_arr[0].axis("off")
ax_arr[1].imshow(rgb_from_hsv)
ax_arr[1]. set_title("RGB_FROM_HSV")
ax_arr[1].axis("off")
plt.show()
```
